# Gradient times Input
This is a very easy to implement analysis, it is implemented in cifar10_utils.py. This is just a demonstration how to use it.

```python
%matplotlib inline
from cifar10_utils import *

net = Net()
net.load("cifar10net")
accuracy = net.accuracy(testloader)
print("Accuracy: ", accuracy)

dataiter = iter(testloader)
images, labels = dataiter.next()

h_itg = input_times_gradient(net, images, one_hot(labels))
plot_heatmaps(images, h_itg.sum(1)*3)
```
